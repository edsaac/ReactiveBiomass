```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
EPS = 1
DAYS_DRY = 1*86400
DAYS_WET = 2*86400

time = range(0, 21*86400+1, 86400//2)
```

```python
timeSteps = list()

for i in time:
    timeSteps.append(i-EPS)
    timeSteps.append(i)
    timeSteps.append(i+EPS)

conc = list()
for i in timeSteps:
    if i%(DAYS_DRY + DAYS_WET) < DAYS_WET:
        conc.append(0.0167)
    else:
        conc.append(0.0001)
    
timeSteps.pop(0)
conc.pop(0)
```

```python
plt.plot(np.array(timeSteps)/86400,conc)
```

```python
stringTable = "(\n"

for c,t in zip(conc,timeSteps):
    stringTable += f"\t\t\t({t:.1f} {c:.4e})\n"

stringTable += ")"
```

```python
with open("stringTableStep.txt",'w') as f:
    f.write(stringTable)
```
