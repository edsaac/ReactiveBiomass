```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use(f'/home/edsaa/Repositories/ReactiveBiomass/misc/edwin.mplstyle')
```

```python
data = pd.read_csv("reviewedProfile.csv")
```

```python
Y = data['Points:1']
data.head()
```

```python
fig,ax = plt.subplots(figsize=[5,6])
fields = ['DOC','O2','NO3','NH4']
ax.plot(data[fields],Y,label=fields)
ax.set(xscale='log')
ax.legend()
plt.show()
```

```python
fig,ax = plt.subplots(figsize=[5,6])
fields = ['rH','rN','rDN']
ax.plot(data[fields],Y,label=fields)
ax.set(xscale='log')
ax.legend()
plt.show()
```

```python
fig,ax = plt.subplots(figsize=[5,6])
ax.fill_betweenx(Y,data['XN'] + data['XDN'] + data['XAR'],x2=0 ,label='XAR')
ax.fill_betweenx(Y,data['XN'] + data['XDN'],x2=0 , label='XN')
ax.fill_betweenx(Y,data['XDN'],x2=0 , label='XDN')
ax.plot(data['O2'],Y)
ax.set(xscale='log')
ax.legend(loc='lower right')
plt.show()
```
